```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2,4,6,8])
y = np.array([81,93, 91, 97])
```

```python
model = Sequential()
```

```python
model.add(Dense(1, input_dim=1, activation='linear'))
# 1은 노드 개수(출력값?), input_dim은 입력 변수 개수(공부 시간), activation은 활성 함수 (출력할 때 어떤 함수로? linear로)
```

```python
model.compile(optimizer='sgd', loss='mse') # sgd -> 랜덤하게(확률적으로) 중간중간 뽑아서 경사 찾아보는 것
```

```python
model.fit(x, y, epochs=2000)
```

```python
plt.scatter(x,y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
hour = 7
prediction = model.predict([hour])
print("%.f 시간을 공부할 경우의 예상 점수는 %.02f 점 입니다." % (hour, prediction))
```
